```python

class Solution:
    def maxArea(self, height: List[int]) -> int:
        area = 0
        for i in range(1, len(height)+1):
            for j in range(1, len(height)+1):
                if (((j - i) * min(height[i], height[j])) >= area):
                    area = (j - i) * (min(height[i], height[j]))
        return area
```

```python
# prompt: how does range in python work

# range(stop)
# This form creates a sequence of integers from 0 to stop - 1.

# range(start, stop)
# This form creates a sequence of integers from start (inclusive) to stop - 1 (exclusive).

# range(start, stop, step)
# This form creates a sequence of integers from start (inclusive) to stop - 1 (exclusive), with a step size of step.

# Examples:

# range(5)
# This will create a sequence of integers from 0 to 4 (inclusive).

# range(1, 5)
# This will create a sequence of integers from 1 to 4 (inclusive).

# range(1, 5, 2)
# This will create a sequence of integers from 1 to 3 (inclusive), with a step size of 2.
```
